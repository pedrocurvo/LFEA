```python
from star import ProtonSTARCalculator, ProtonMaterials

import matplotlib.pyplot as plt
import numpy as np

material1 = ProtonMaterials.SILICON 
material2 = ProtonMaterials.IRON 
material3 = ProtonMaterials.POLYETHYLENE
#ALUMINUM; IRON; POLYETHILENE; SILICON
calculator1 = ProtonSTARCalculator(material1)
calculator2 = ProtonSTARCalculator(material2)
calculator3 = ProtonSTARCalculator(material3)

plt.style.use('_mpl-gallery')
# make data
x = np.logspace(-3, 4, 100000)
y1 = calculator1.calculate_total_stopping_powers(x) # Energy in MeV
y2 = calculator2.calculate_total_stopping_powers(x) # Energy in MeV
y3 = calculator3.calculate_total_stopping_powers(x) # Energy in MeV

# plot
fig, ax = plt.subplots(1, figsize=(10,8))

ax.loglog(x, y1, linewidth=2.0)
ax.loglog(x, y2, linewidth=2.0)
ax.loglog(x, y3, linewidth=2.0)



#plt.yscale("log")
#plt.xscale("log")
#ax.set(xlim=(0.1, 10000), xticks=np.arange(1, 10000),
#       ylim=(0.1, 1000), yticks=np.arange(1, 8))
#ax.set(xlim=(0.001, 10000), xticks=np.linspace(1, 10000),
#       ylim=(0, 5000), yticks=np.logspace(0, 4))

ax.set_title("Proton Stopping power for "+ material.name)
ax.set_ylabel("dE/dx [g/cm^2]")
ax.set_xlabel("E [Mev]")
plt.grid(which='minor')
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
t = np.arange(0.01, 20.0, 0.01)

# Create figure
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

# log y axis
ax1.semilogy(t, np.exp(-t / 5.0))
ax1.set(title='semilogy')
ax1.grid()

# log x axis
ax2.semilogx(t, np.sin(2 * np.pi * t))
ax2.set(title='semilogx')
ax2.grid()

# log x and y axis
ax3.loglog(t, 20 * np.exp(-t / 10.0))
ax3.set_xscale('log', base=2)
ax3.set(title='loglog base 2 on x')
ax3.grid()

# With errorbars: clip non-positive values
# Use new data for plotting
x = 10.0**np.linspace(0.0, 2.0, 20)
y = x**2.0

ax4.set_xscale("log", nonpositive='clip')
ax4.set_yscale("log", nonpositive='clip')
ax4.set(title='Errorbars go negative')
ax4.errorbar(x, y, xerr=0.1 * x, yerr=5.0 + 0.75 * y)
# ylim must be set after errorbar to allow errorbar to autoscale limits
ax4.set_ylim(bottom=0.1)

fig.tight_layout()
plt.show()
```
